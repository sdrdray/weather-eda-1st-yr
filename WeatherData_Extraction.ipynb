```python
URL = "http://api.weatherapi.com/v1/history.json?key=5ffe7d6a58804d74af7162902242305&q={}&dt={}"

STATES_CITIES = {'Andhra Pradesh' : ["Visakhapatnam", "Vijayawada", "Guntur", "Nellore", "Kurnool", "Rajahmundry", "Tirupati", "Kakinada", "Kadapa",
                                     "Anantapur", "Vizianagaram", "Eluru", "Ongole", "Nandyal", "Machilipatnam", "Adoni", "Tenali", "Proddatur", 
                                     "Chittoor", "Hindupur", "Bhimavaram", "Madanapalle", "Guntakal", "Srikakulam", "Dharmavaram", "Gudivada",
                                     "Narasaraopet", "Tadipatri", "Tadepalligudem", "Amaravati", "Chilakaluripet", "Palakollu", "Kavali", "Sullurpeta",
                                     "Bapatla", "Nandyal", "Markapur", "Kandukur", "Rayachoti", "Giddalur", "Narasapuram", "Narasannapeta", 
                                     "Pithapuram", "Parvathipuram", "Chirala", "Vinukonda", "Punganur", "Sattenapalle", "Jaggaiahpet", "Tuni",
                                     "Repalle", "Nuzvid", "Peddapuram", "Yemmiganur", "Dhone", "Mandapeta", "Venkatagiri"],
                 
                 
                 
      'Rajasthan': [
        "Jaipur", "Jodhpur", "Kota", "Bikaner", "Bhiwadi",
        "Alwar", "Udaipur", "Ajmer", "Bhilwara", "Sikar",
        "Bharatpur", "Pali", "Sri Ganganagar", "Beawar", "Baran",
        "Dhaulpur", "Tonk", "Kishangarh", "Hanumangarh", "Banswara",
        "Dungarpur", "Pratapgarh", "Chittorgarh", "Jhalawar", "Nagaur",
        "Sawai Madhopur", "Barmer", "Churu", "Jalore", "Sirohi",
        "Bhinmal", "Sardarshahr", "Nathdwara", "Rawatbhata", "Rajgarh",
        "Dholpur", "Suratgarh", "Fatehpur", "Abu Road", "Lachhmangarh",
        "Sojat", "Neem-Ka-Thana", "Sujangarh", "Pilibanga", "Shahpura",
        "Makrana", "Nohar", "Mandawa", "Kotputli", "Ramganj Mandi",
        "Kuchaman City", "Raisinghnagar", "Manasa", "Losal", "Nasirabad",
        "Sri Madhopur", "Nokha", "Sadri", "Rajaldesar", "Bhadesar",
        "Rajakhera", "Mundwa", "Todaraisingh"
    ],
    'Goa': [
        "Panaji", "Margao", "Vasco da Gama", "Mapusa", "Ponda",
        "Calangute", "Palolem Beach", "Anjuna", "Candolim", "Silvassa"
    ],
    'Assam': [
        "Guwahati", "Silchar", "Dibrugarh", "Jorhat", "Nagaon",
        "Tinsukia", "Tezpur", "Goalpara", "Dhubri", "Sivasagar",
        "Hailakandi", "Lakhimpur", "Sonitpur", "Cachar", "Karimganj",
        "Darrang", "Nalbari", "Bongaigaon", "Kokrajhar", "Dhemaji",
        "Karbi Anglong", "Golaghat", "Barpeta", "Kamrup", "Morigaon",
        "Marigaon", "Dima Hasao", "Hojai", "Udalguri", "Chirang",
        "Baksa"
    ],
    'Arunachal Pradesh': [
        "Itanagar", "Naharlagun", "Khonsa", "Roing", "Ziro Valley",
        "Tawang", "Changlang", "Deban", "Dharampur", "Jairampur",
        "Kharsang", "Khemiyong", "Kherem Bisa", "Kutum", "Manabhum",
        "Manmao", "Miao", "Namchik", "Namdang", "Namphai"
    ] ,
    'Manipur': [
        "Imphal", "Thoubal", "Bishnupur", "Senapati", "Chandel",
        "Tamenglong", "Ukhrul", "Churachandpur", "Imphal East", "Imphal West",
        "Kakching", "Jiribam", "Kangpokpi", "Noney", "Pherzawl", "Kamjong",
        "Tengnoupal"
    ],
    'Meghalaya': [
        "Shillong", "Cherrapunji", "Jowai", "Tura", "Nongstoin",
        "East Garo Hills", "West Garo Hills", "Mawlynnong", "Williamnagar",
        "Balpakram National Park", "Elephant Falls", "Double-decker Living Root Bridge",
        "Don Bosco Museum", "Kyllang Rock", "Mawlynnong Village", "Baghmara",
        "Nohkalikai Falls", "Mawsmai Cave", "Umiam Lake", "Dawki", "Laitlum Canyon",
        "Mawsynram", "Mawphlang Sacred Forest", "Sohra", "Nongpoh", "Ranikor"
    ],
    'Mizoram': [
        "Aizawl", "Serchhip", "Lunglei", "Champhai", "Mamit",
        "Kolasib", "Falkawn Village", "Saiha", "Lawngtlai", "Saitual",
        "Khawzawl", "Hnahthial", "Sialsuk", "Thenzawl", "Vairengte",
        "Thingsulthliah", "Keifang", "Neihbawi"
    ],
    'Nagaland': [
        "Kohima", "Dimapur", "Khonoma", "Mon", "Benreu",
        "Phek", "Mokokchung", "Wokha", "Tuensang", "Zunheboto"
    ],
    'Tripura': [
        "Agartala", "Dharmanagar", "Udaipur", "Kailashahar", "Bishalgarh",
        "Teliamura", "Khowai", "Belonia", "Melaghar", "Mohanpur", "Ambassa",
        "Ranirbazar", "Santirbazar", "Kumarghat", "Sonamura", "Panisagar",
        "Amarpur", "Jirania", "Kamalpur", "Sabroom", "Khayerpur", "Kanchanpur",
        "Bishramganj", "Manu", "Mungiakami", "Rupaichari", "Taidu", "Birganj",
        "Dumburnagar", "Gakulnagar", "Jolaibari", "Chandrapur", "Kakraban",
        "Kalacherra"
    ],
    'Sikkim': [
        "Gangtok", "Pelling", "Lachung", "Nathula Pass", "MG Marg",
        "Gurudongmar Lake", "Buddha Park", "Banjhakri Water Falls", "Ranka Monastery",
        "Hanuman Tok", "Tashi View Point", "Namgyal Institute of Tibetology", "Do Drul Chorten Stupa",
        "Enchey Monastery", "Ganesh Tok", "Saramsa Garden", "Phodong Monastery", "Kabi Longstok",
        "Singhik", "Chungthang", "Lachen", "Thangu", "Mangan", "Zuluk", "Aritar"
    ],
    'Kerala': [
        "Thiruvananthapuram", "Kochi", "Kozhikode", "Kollam", "Thrissur",
        "Kannur", "Alappuzha", "Palakkad", "Malappuram", "Manjeri",
        "Thalassery", "Ponnani", "Vatakara", "Kanhangad", "Payyanur",
        "Koyilandy", "Parappanangadi", "Kalamassery", "Kodungallur", "Neyyattinkara",
        "Tanur", "Kayamkulam", "Guruvayur", "Thrikkakkara", "Wadakkancherry",
        "Nedumangad", "Kondotty", "Tirurangadi", "Tirur", "Chalakudy",
        "Payyoli", "Koduvally", "Mananthavady", "Changanassery", "Mattanur",
        "Punalur", "Nilambur", "Chennamangalam", "Anakkayam", "Dharmadam"
    ],
    'Gujarat': [
        "Ahmedabad", "Surat", "Vadodara", "Rajkot", "Bhavnagar",
        "Jamnagar", "Gandhinagar", "Junagadh", "Gandhidham", "Anand",
        "Navsari", "Morbi", "Nadiad", "Surendranagar", "Bharuch",
        "Mehsana", "Bhuj", "Porbandar", "Palanpur", "Valsad",
        "Vapi", "Gondal", "Veraval", "Godhra", "Patan",
        "Kalol", "Dahod", "Botad", "Amreli", "Deesa",
        "Jetpur", "Anjar", "Kodinar", "Visnagar", "Dhoraji",
        "Kapadvanj", "Mandvi", "Palitana", "Umreth", "Savarkundla",
        "Talaja", "Padra", "Mahesana", "Keshod", "Wadhwan",
        "Modasa", "Lunawada", "Mangrol", "Viramgam", "Sanand",
        "Sidhpur", "Kadi", "Wankaner", "Sihor", "Mahemdabad",
        "Khambhat", "Visavadar", "Limbdi", "Songadh", "Radhanpur",
        "Vijapur", "Ranavav", "Thangadh", "Halvad", "Rapar",
        "Una", "Ankleshwar", "Rajpipla", "Vyara", "Idar",
        "Bhujodi", "Vankaner", "Lathi", "Jasdan"
    ],
    'Maharashtra': [
        "Mumbai", "Pune", "Nagpur", "Thane", "Pimpri Chinchwad",
        "Nashik", "Kalyan Dombivli", "Vasai Virar", "Aurangabad", "Navi Mumbai",
        "Solapur", "Mira Bhayandar", "Bhiwandi Nizampur", "Jalgaon", "Amravati",
        "Nanded Waghala", "Kolhapur", "Ulhasnagar", "Sangli Miraj Kupwad", "Malegaon",
        "Akola", "Latur", "Dhule", "Ahmednagar", "Chandrapur",
        "Parbhani", "Ichalkaranji", "Jalna", "Ambarnath", "Bhusawal",
        "Panvel", "Badlapur", "Beed", "Gondia", "Satara",
        "Barshi", "Yavatmal", "Achalpur", "Osmanabad", "Nandurbar",
        "Wardha", "Udgir", "Hinganghat", "Ratnagiri", "Sangli",
        "Nanded", "Ambarnath", "Vasai-Virar", "Baramati", "Alibag",
        "Lonavala", "Karad", "Pen", "Palghar", "Dhulia",
        "Mahad", "Phaltan", "Rahuri", "Palus", "Shegaon",
        "Shirur", "Sinnar", "Warora", "Tuljapur", "Lonar",
        "Chalisgaon", "Sangamner", "Manmad", "Chandvad", "Shahada",
        "Manwath", "Morshi", "Mul", "Manchar", "Manjlegaon",
        "Mouda", "Mudhol", "Mukhed", "Muktainagar", "Murum"
    ],
    'Odisha': [
        "Bhubaneswar", "Cuttack", "Rourkela", "Berhampur", "Sambalpur",
        "Puri", "Balasore", "Bhadrak", "Baripada", "Balangir",
        "Jeypore", "Keonjhar", "Bhawanipatna", "Chandipur", "Dhauli",
        "Dhenkanal", "Gopalpur", "Lalitgiri", "Udayagiri", "Ratnagiri",
        "Pipli", "Raghurajpur", "Satyabadi", "Taptapani", "Konark",
        "Khordha", "Sundargarh", "Ganjam", "Mayurbhanj"
    ],
    'West Bengal': [
        "Kolkata", "Asansol", "Siliguri", "Durgapur", "Bardhaman",
        "Malda", "Baharampur", "Habra", "Kharagpur", "Shantipur",
        "Dankuni", "Dhulian", "Ranaghat", "Haldia", "Raiganj",
        "Krishnanagar", "Nabadwip", "Medinipur", "Jalpaiguri", "Balurghat",
        "Basirhat", "Bankura", "Chakdaha", "Darjeeling", "Alipurduar",
        "Purulia", "Jangipur", "Bolpur", "Bangaon", "Cooch Behar",
        "Howrah", "Rajpur Sonarpur", "Bidhannagar", "Dum Dum", "Barasat",
        "Bally", "Kamarhati", "Panihati", "Kalyani", "Barrackpore",
        "Bhatpara", "Serampore", "North Dumdum", "South Dumdum", "Baranagar",
        "Uluberia", "Naihati", "Madhyamgram", "Kanchrapara", "Ashoknagar Kalyangarh",
        "Rajarhat Gopalpur", "Bidhan Nagar"
    ],
    'Karnataka': [
        "Bangalore", "Mysore", "Hubballi-Dharwad", "Kalaburagi", "Mangalore",
        "Belagavi", "Davanagere", "Bellary", "Vijayapura", "Shimoga",
        "Tumkur", "Raichur", "Bidar", "Udupi", "Hospet",
        "Gadag-Betageri", "Robertsonpet", "Hassan", "Bhadravati", "Chitradurga",
        "Kolar", "Mandya", "Chikmagalur", "Gangavati", "Bagalkot",
        "Ranebennuru", "Arsikere", "Gulbarga", "Bijapur", "Tumakuru",
        "Chikkamagaluru", "Dakshina Kannada", "Kodagu", "Uttara Kannada", "Chamarajanagar",
        "Ramanagara", "Yadgir", "Haveri", "Chikkaballapura", "Koppal",
        "Ballari", "Bangalore Rural"
    ],
    'Andhra Pradesh': [
        "Visakhapatnam", "Vijayawada", "Guntur", "Nellore", "Kurnool",
        "Kakinada", "Rajamahendravaram", "Kadapa", "Mangalagiri-Tadepalli", "Tirupati",
        "Anantapuram", "Vizianagaram", "Ongole", "Eluru", "Proddatur",
        "Nandyal", "Adoni", "Madanapalle", "Machilipatnam", "Tenali",
        "Chittoor", "Hindupur", "Srikakulam", "Bhimavaram", "Tadepalligudem",
        "Guntakal", "Dharmavaram", "Gudivada", "Narasaraopet", "Kadiri",
        "Tadipatri", "Chilakaluripet", "Amaravati", "Araku Valley", "Rajahmundry",
        "Srikalahasti", "Chirala", "Anakapalle", "Tuni", "Amalapuram",
        "Pithapuram", "Nagari", "Palakollu", "Sattenapalle", "Yemmiganur",
        "Kanigiri", "Kandukur", "Parvathipuram", "Bobbili", "Ponnur",
        "Narasapuram", "Tanuku", "Rayachoti", "Srisailam"
    ],
    'Telangana': [
        "Hyderabad", "Warangal", "Nizamabad", "Khammam", "Karimnagar",
        "Ramagundam", "Mahabubnagar", "Nalgonda", "Adilabad", "Suryapet",
        "Miryalaguda", "Siddipet", "Jagtial", "Mancherial", "Armoor",
        "Nirmal", "Sircilla", "Kamareddy", "Palwancha", "Kothagudem",
        "Bodhan", "Sangareddy", "Metpally", "Zahirabad", "Meerpet Jillelguda",
        "Medchal Malkajgiri", "Ranga Reddy", "Medak", "Peddapalli"
    ],
    'Tamil Nadu': [
        "Chennai", "Coimbatore", "Madurai", "Tiruchirapalli", "Tiruppur",
        "Salem", "Erode", "Tirunelveli", "Vellore", "Thoothukudi",
        "Dindigul", "Thanjavur", "Nagercoil", "Kanchipuram", "Karaikkudi",
        "Neyveli", "Cuddalore", "Kumbakonam", "Tiruvannamalai", "Rajapalayam",
        "Pudukkottai", "Ambur", "Nagapattinam", "Hosur", "Tambaram",
        "Alandur", "Pallavaram", "Avadi", "Tiruvottiyur", "Ambattur",
        "Kurichi", "Madavaram", "Sivagangai", "Krishnagiri", "Tirupattur",
        "Virudhunagar"
    ],
    'Uttarakhand': [
        "Dehradun", "Haridwar", "Roorkee", "Haldwani", "Rudrapur",
        "Kashipur", "Rishikesh", "Mussoorie", "Nainital", "Almora",
        "Khatima", "Kotdwar", "Pantnagar", "Pauri", "Ramnagar",
        "Ranikhet", "Srinagar Garhwal", "Uttarkashi", "Vikas Nagar", 
        "Bageshwar", "Chamoli", "Champawat", "Gopeshwar", "Joshimath",
        "Kichha", "Lansdowne", "Manglaur", "Pithoragarh", "Tehri",
        "Bhimtal", "Chakrata", "Dwarahat", "Gangotri", "Jaspur",
        "Kedarnath", "Laksar", "Muni Ki Reti", "Nagina", "Purola",
        "Sitarganj", "Tanakpur", "Badrinath", "Barkot", "Devprayag",
        "Dharchula", "Dogadda", "Doiwala", "Gadarpur", "Gularbhoj",
        "Herbertpur", "Jhabrera", "Karnaprayag", "Kela Khera", "Landhaura",
        "Lohaghat", "Mahua Dabra Haripura", "Mahua Kheraganj", "Nandprayag"
    ],
    'Haryana': [
        "Faridabad", "Fatehabad", "Ganaur", "Gharaunda", "Gohana",
        "Gurugram", "Hansi", "Hisar", "Hodal", "Indri",
        "Jagadhri", "Jhajjar", "Jind", "Julana", "Kaithal",
        "Kalanwali", "Kalka", "Kanina", "Karnal", "Kharkhoda",
        "Ladwa", "Maham", "Mahendragarh", "Mandi Dabwali", "Mustafabad",
        "Naraingarh", "Narnaul", "Narwana", "Nilokheri", "Nuh",
        "Meham", "Palwal", "Panchkula", "Panipat", "Pehowa",
        "Pinjore", "Pundri", "Rania", "Ratia", "Rewari",
        "Rohtak", "Safidon", "Samalkha", "Shahbad", "Sirsa",
        "Sohna", "Sonipat", "Taraori", "Thanesar", "Tohana",
        "Yamunanagar", "Abohar", "Adampur", "Ahmedgarh", "Ajnala",
        "Amloh", "Amritsar", "Bagha Purana", "Balachaur", "Batala",
        "Bathinda", "Kalanaur", "Kalka", "Karnal", "Kharkhoda",
        "Kosli", "Ladwa", "Maham", "Mahendragarh", "Mandi Dabwali",
        "Narnaul", "Narwana", "Nilokheri", "Nuh", "Palwal",
        "Pinjore", "Pundri", "Rania", "Ratia", "Rewari",
        "Rohtak", "Samalkha", "Shahbad", "Sirsa", "Sohna",
        "Sonipat", "Taraori", "Thanesar", "Tohana", "Yamunanagar",
        "Abohar", "Adampur", "Ahmedgarh", "Ajnala"
    ],
    'Bihar': [
        "Patna", "Gaya", "Bhagalpur", "Muzaffarpur", "Darbhanga",
        "Purnia", "Sasaram", "Arrah", "Samastipur", "Begusarai",
        "Katihar", "Bettiah", "Motihari", "Saharsa", "Munger",
        "Chhapra", "Siwan", "Madhubani", "Hajipur", "Buxar",
        "Sitamarhi", "Kishanganj", "Rajgir", "Nalanda", "Nawada",
        "Khagaul", "Saidpura", "Danapur", "Dinapur", "Kaler",
        "Paharpur", "Habibpur", "Sabour", "Nathnagar", "Damodarpur",
        "Dumari", "Majhauli", "Yehiapur", "Shekhpura", "Kasba",
        "Padri", "Bahadurpur", "Sanrha", "Mansaraut", "Kargahia",
        "Saraiya", "Bhadauni", "Sarimpur", "Rajopatti", "Talkhapur",
        "Dumra", "Forbesganj", "Sultanganj", "Jamalpur", "Barh",
        "Lakhisarai", "Sheohar", "Supaul", "Araria"
    ],
    'Jharkhand': [
        "Jamshedpur", "Dhanbad", "Ranchi", "Bokaro", "Deoghar",
        "Phusro", "Hazaribagh", "Giridih", "Ramgarh", "Medininagar",
        "Chirkunda", "Dumka", "Chaibasa", "Sahibganj", "Pakur",
        "Koderma", "Lohardaga", "Gumla", "Simdega", "Chatra",
        "Garhwa", "Khunti", "Godda", "Latehar", "Saraikela",
        "Jamtara", "Palamu", "West Singhbhum", "East Singhbhum"
    ],
    'Jammu and Kashmir ': [
        "Srinagar", "Jammu", "Anantnag", "Baramulla", "Sopore",
        "Kathua", "Udhampur", "Rajouri", "Poonch", "Leh",
        "Kargil", "Kishtwar", "Doda", "Akhnoor", "Bandipora",
        "Ganderbal", "Kulgam", "Pulwama", "Shopian", "Budgam",
        "Samba", "Pattan", "Kupwara", "Banihal", "Ramban",
        "Gulmarg", "Pahalgam", "Bhaderwah", "Reasi", "Billawar",
        "Hiranagar", "Katra", "Achabal", "Chenani", "Handwara",
        "Kreeri", "Magam", "Shangus", "Sumbal", "Tral"
    ],
    'Himachal Pradesh ': [
        "Shimla", "Dharamsala", "Solan", "Mandi", "Palampur",
        "Baddi", "Nahan", "Paonta Sahib", "Sundarnagar", "Chamba",
        "Una", "Kullu", "Hamirpur", "Bilaspur", "Yol",
        "Nalagarh", "Nurpur", "Kangra", "Baijnath Paprola", "Santokhgarh",
        "Mehatpur Basdehra", "Shamshi", "Parwanoo", "Manali", "Tira Sujanpur",
        "Ghumarwin", "Dalhousie", "Rohru", "Nagrota Bagwan", "Rampur",
        "Jawalamukhi", "Jogindernagar", "Dera Gopipur", "Sarkaghat", "Jhakhri",
        "Indora", "Bhuntar", "Nadaun", "Theog", "Kasauli",
        "Gagret", "Chuari Khas", "Manikaran", "Keylong", "Reckong Peo",
        "Kaza", "Tabo", "Kalpa", "Rajgarh", "Arki",
        "Kotkhai", "Jubbal", "Seoni", "Banjar", "Karsog",
        "Sarka Ghat", "Pachhad", "Sangla", "Chitkul", "Pooh"
    ],
    'Punjab ': [
        "Ludhiana", "Amritsar", "Jalandhar", "Patiala", "Bathinda",
        "Hoshiarpur", "Mohali", "Batala", "Pathankot", "Moga",
        "Abohar", "Malerkotla", "Khanna", "Phagwara", "Muktsar",
        "Barnala", "Rajpura", "Firozpur", "Kapurthala", "Sunam",
        "Gurdaspur", "Tarn Taran", "Sangrur", "Nabha", "Talwandi Sabo",
        "Jagraon", "Mansa", "Kotkapura", "Sri Muktsar Sahib", "Faridkot",
        "Dhuri", "Gidderbaha", "Jaitu", "Dasuya", "Mukerian",
        "Nangal", "Fazilka", "Zira", "Kot Isa Khan", "Samana",
        "Sardulgarh", "Bhawanigarh", "Payal", "Raikot", "Longowal",
        "Dera Bassi", "Khamano", "Morinda", "Nawanshahr", "Machhiwara",
        "Patti", "Rupnagar", "Bhikhi", "Bhikhiwind", "Balachaur",
        "Dhanaula", "Gobindgarh", "Jalalabad", "Phillaur", "Qadian",
        "Raman", "Tappa", "Budhlada", "Garhshankar", "Ghanaur",
        "Goraya", "Kharar", "Lehragaga", "Mahilpur", "Malerkotla"
    ],
    'Uttar Pradesh' : ["Kanpur", "Lucknow", "Ghaziabad", "Agra", "Meerut", "Varanasi", "Prayagraj", "Bareilly", "Aligarh", "Moradabad", "Saharanpur",
             "Gorakhpur", "Noida", "Firozabad", "Loni", "Jhansi", "Muzaffarnagar", "Mathura", "Shahjahanpur", "Rampur", "Farrukhabad",
             "Maunath Bhanjan", "Hapur", "Etawah", "Mirzapur", "Bulandshahr", "Sambhal", "Amroha", "Hardoi", "Fatehpur", "Raebareli",
             "Orai", "Sitapur", "Bahraich", "Modinagar", "Unnao", "Jaunpur", "Lakhimpur", "Hathras", "Banda", "Pilibhit", "Mughalsarai",
             "Barabanki", "Khurja", "Gonda", "Mainpuri", "Lalitpur", "Etah", "Deoria", "Ujhani", "Ghazipur", "Sultanpur", "Azamgarh",
             "Sahaswan", "Basti", "Chandausi", "Akbarpur", "Ballia", "Tanda", "Greater Noida", "Shikohabad", "Shamli", "Khair", "Kasganj",
             "Bijnor", "Mau", "Fatehpur Sikri", "Amethi", "Gyanpur", "Khalilabad", "Robertsganj", "Gopiganj", "Gola Gokarannath", "Tundla",
             "Colonelganj", "Mahoba", "Chunar", "Nawabganj", "Puranpur", "Biswan", "Chakia", "Utraula", "Ghosi", "Rasra", "Sikandra Rao",
             "Bilariaganj", "Powayan", "Mihinpurwa", "Birdpur", "Chitrakoot", "Itaunja", "Pihani", "Bansi", "Atarra", "Baberu", "Nanpara",
             "Kunda", "Chirgaon", "Bharthana", "Bisauli", "Bighapur", "Lalganj", "Pawayan", "Jhinjhak", "Bikapur", "Meerut Cantt", "Handia",
             "Kandhla"]

                 }

date_format = ["2024-5-{}".format(i) for i in range(13, 14)]

def weather_API(STATES_CITIES, URL):

  # Dataframe to store location
  location_df = pd.DataFrame(columns = ['name', 'region', 'country', 'lat', 'lon', 'tz_id', 'localtime_epoch', 'localtime'])
  # Dataframe to store forecasts
  forecast_df = pd.DataFrame(columns = ['time_epoch', 'time', 'temp_c', 'temp_f', 'is_day', 'condition',
                                        'wind_mph', 'wind_kph', 'wind_degree', 'wind_dir', 'pressure_mb',
                                        'pressure_in', 'precip_mm', 'precip_in', 'humidity', 'cloud',
                                        'feelslike_c', 'feelslike_f', 'windchill_c', 'windchill_f',
                                        'heatindex_c', 'heatindex_f', 'dewpoint_c', 'dewpoint_f',
                                        'will_it_rain', 'chance_of_rain', 'will_it_snow', 'chance_of_snow',
                                        'vis_km', 'vis_miles', 'gust_mph', 'gust_kph', 'state','city'])
  # Dataframe to store astro
  astro_df = pd.DataFrame(columns = ['sunrise', 'sunset', 'moonrise', 'moonset', 'moon_phase', 'moon_illumination', 'state', 'city'])
    
  # iterate over states
  for state in STATES_CITIES.keys():
    print("Extracting data for state {}".format(state))
    print()

    # iterate over all cities
    cities = STATES_CITIES[state]
    state_arr = np.array([state for _ in range(24)])

    for j, date in enumerate(date_format):

        print('Date : {}'.format(date))
        city_count = 0

        for i, city in enumerate(cities):

          print("Extracting data for city {}".format(city))
          print()

          # Create the final URL
          final_url = URL.format(city, date)

          # make get request
          try:
            time.sleep(10)
            page_response = requests.get(final_url)
          except:
            print("Could not make GET request for state : {} city : {} and date : {}".format(city, state, date))

          # extract the page_response content
          data_dict = eval(page_response.content)


          try:
            # location_data
            location_data = data_dict['location']
            curr_loc_df = pd.DataFrame(data = location_data, index = [i])

            # concat with main frame
            location_df = pd.concat((location_df, curr_loc_df), axis=0)
          except:

            print("Could not extract location data for state : {} city : {} date: {}".format(state, city, date))

          try:
            # forecast_data
            forecast_data = data_dict['forecast']['forecastday'][0]['hour']
            curr_forecast_df = pd.DataFrame(forecast_data)
            num_rows = curr_forecast_df.shape[0]

            # Create city arr
            city_arr = np.array([city for _ in range(num_rows)])

            # Add state array and city array in current forecast df
            curr_forecast_df['city'] = city_arr
            curr_forecast_df['state'] = state_arr

            # concat with main frame
            forecast_df = pd.concat((forecast_df, curr_forecast_df), axis=0)
          except:
            print("Could not extract forecast data for state : {} city : {} data : {}".format(state, city, date))
          
          try:
            # Astro data
            astro_data = data_dict['forecast']['forecastday'][0]['astro']
            curr_astro_df = pd.DataFrame(astro_data, index=[id])
            curr_astro_df['city'] = [city]
            curr_astro_df['state'] = [state]

            # Concat with main frame
            astro_df = pd.concat((astro_df, curr_astro_df), axis=0)
          except:
            print("Could not extract astro data for state : {} city : {} date : {}".format(state, city, date))
            city_count -= 1
          
            city_count += 1
        print("Extracted {} out of {} cities for state {}".format(city_count, len(cities), state))
        print()

  return location_df, forecast_df, astro_df
```
